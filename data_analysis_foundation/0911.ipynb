### 정렬이 여러 개 일 때

- 두 개 이상의 조건으로 차순을 정리

- 두 개 이상일 때는 2차원의 개념으로 들어가서 반드시 리스트로 작성해 주어야 함

```python
## 두 개 이상 리스트로 작성

df.sort_values(['english','science'])

## english(=main) 오름차순으로 정렬되고 , science는 서브 / english의 정렬에 따라 움직임
# english의 동일 한 68점을 보면 science가 87 다음 98 로 오름차순으로 잘 정렬된 것을 볼 수 있음
```

```python
df.sort_values(['english','nclass'], ascending=[True,False])

# english는 오름차순이고 nclass는 내림차순이니, english의 78점을 보면 nclass가 더 큰 5 다음 2가 오는 것을 볼 수 있다.
```

### 파생변수 만들기

- 파생변수 : 내부 데이터를 가지고 새로운 변수를 만드는 것 
- 왜? 
1. 데이터 분석에 필요한 경우 ex) 새로운 평균이라는 기준으로 데이터 분석하는 경우
2. 모델링을 통해 피처에 대한 엔지니어링을 하는 경우

     우리가 해결하고자 하는 문제를 컴퓨터가 잘 이해할 수 있도록 변수(피처)들의 형태를 변형하거나 적절하게 처리하는데, 
     이러한 과정을 피처 엔지니어링 이라고 함.

- how?

1. 컬럼+컬럼으로 바로 사용
2. apply 함수로 lambda 적용
3. loc, iloc 사용해서
4. assign() 함수 사용해서

```python
df['2_avg']= (df['english']+df['math'])/2
```

```python
df.assign(avg = (df['english']+df['math'])/2)
```

### assign 함수 이용해서 여러가지 파생변수 만들기

```python
df['2_avg'].mean() 
# 개인 평균의 평균
```

```python
df.assign(avg = (df['english']+df['math'])/2, inplace=True).head(7)

# 평균 구하고 원본 데이터 수정 한 다음 앞에 7개 만 출력
```

```python
df
 # 원본 데이터에 2_avg가 추가된 것을 볼 수 있음
```

```python
import numpy as np
```

```python
df.assign(result= np.where(df['2_avg']>66,'평균이상','평균이하'))
```

```python
np.where(df['2_avg']>66)

## 넘파이 함수 -> index로 위치 잡고, 그 위치에 값을 넣음 
```

### groupby

- 그룹으로 묶어 통계치 계산, 카운팅 계산
- df.groupby(기준).agg(변수,요약통계치)결과를 확인

```python
# groupby는 agg 함수와 같이 많이 사용
df.groupby('nclass')
```

```python
df.groupby('nclass').agg(english_avg = ('english','sum'))
# nclass별 영어 평균
```

```python
df.groupby('nclass').agg(S_sum = ('science','sum'))
```
